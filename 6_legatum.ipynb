### Legatum Prosperity Index 2023 to Data Base

```python
from os import getenv

import pandas as pd
from sqlalchemy import create_engine, text
```

```python
engine = create_engine(getenv("SQLALCHEMY_RELOHELPER_URL"))
```

```python
df_prosperity = pd.read_csv('./data/Dataset_Legatum_Prosperity_Index_2023.csv', sep=';')
df_prosperity.rename(columns={'area_code': 'country_code'}, inplace=True)
df_prosperity.to_sql('legatum_country_indices', engine, if_exists='append', index=False)

stmt6 = "ALTER TABLE legatum_country_indices ADD CONSTRAINT PK_lagatum_index PRIMARY KEY (country_code, pillar_name)"
stmt7 = "ALTER TABLE legatum_country_indices ADD CONSTRAINT FK_country_code FOREIGN KEY (country_code) REFERENCES public.countries (country_code)"

with engine.connect() as connection:
    connection.execute(text(stmt6))
    connection.execute(text(stmt7))
    connection.commit()
```
