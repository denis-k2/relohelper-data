Приведение `geonameid` и `population` к типу `Int64`

```python
df["geonameid"] = df["geonameid"].astype("Int64")
df["population"] = df["population"].astype("Int64")
```

```python
df
```

Из главного большого датафрейма формирую 3 таблицы для их последующей отправки в БД

```python
audit_date = date.today()
audit_by = getenv("NB_USER")

df_numbeo_city_indices = df.loc[
    :,
    [
        "geonameid",
        "cost_of_living",
        "rent",
        "cost_of_living_plus_rent",
        "groceries",
        "local_purchasing_power",
        "quality_of_life",
        "property_price_to_income_ratio",
        "traffic_commute_time",
        "climate",
        "safety",
        "health_care",
        "pollution",
    ],
].copy()

df_cities = df.loc[
    :,
    [
        "geonameid",
        "city",
        "state_code",
        "country_code",
        "population",
        "latitude",
        "longitude",
        "timezone",
    ],
].copy()

df_numbeo_city_links = df.loc[
    :,
    [
        "geonameid",
        "link",
    ],
].copy()

for d in (df_numbeo_city_indices, df_cities, df_numbeo_city_links):
    d.loc[:, "updated_date"] = audit_date
    d.loc[:, "updated_by"] = audit_by
```

```python
df_numbeo_city_links = df_numbeo_city_links.rename(columns={"link": "numbeo_link"})
df_numbeo_city_links.insert(2, "climate_link", None)
df_numbeo_city_links.head(3)
```

```python

df_numbeo_city_indices.to_sql(
    "numbeo_city_indices", engine, if_exists="append", index=True
)

df_cities.to_sql("cities", engine, if_exists="append", index=True)

df_numbeo_city_links.to_sql("city_source_links", engine, if_exists="append", index=True)
```

```python
stmt = """
ALTER TABLE public.cities
ADD CONSTRAINT cities_pkey PRIMARY KEY (city_id)
"""

with engine.connect() as connection:
    connection.execute(text(stmt))
    connection.commit()
```

```python
stmt1 = """
ALTER TABLE public.cities
ADD CONSTRAINT FK_cities_state
FOREIGN KEY (state_code)
REFERENCES public.state (state_code)
"""

stmt2 = """
ALTER TABLE public.cities
ADD CONSTRAINT FK_cities_country
FOREIGN KEY (country_code)
REFERENCES public.countries (country_code)
"""

with engine.connect() as connection:
    connection.execute(text(stmt1))
    connection.execute(text(stmt2))
    connection.commit()
```
